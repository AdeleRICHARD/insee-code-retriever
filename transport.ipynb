# Fonction qui permet d'appeler l'api sncf et de récupérer les communes associées aux stations

```python
def is_in_idf(insee_code):
    """
    Vérifie si un code INSEE appartient à l'Île-de-France.
    """
    return insee_code.startswith(('75', '77', '78', '91', '92', '93', '94', '95'))

def get_commune_from_station_sncg(station_name, api_key, transport_mode):
    """
    Recherche la commune d'une gare via l'API SNCF en utilisant la clé API comme nom d'utilisateur.
    
    station_name: Nom de la gare
    api_key: Clé API SNCF (utilisée comme nom d'utilisateur)
    transport_mode: Type de transport (ex: GL, METRO, TRAMWAY, etc.)
    """
    url = f"https://api.sncf.com/v1/coverage/sncf/places?q={station_name}"
    response = requests.get(url, auth=HTTPBasicAuth(api_key, ''))  # Le champ mot de passe est vide
    
    if response.status_code == 200:
        data = response.json()
        if 'places' in data and len(data['places']) > 0:
            for place in data['places']:
                if 'stop_area' in place:
                    stop_area = place['stop_area']
                    if 'administrative_regions' in stop_area:
                        admin_region = stop_area['administrative_regions'][0]
                        insee_code = admin_region['insee']
                        
                        # Si c'est un transport IDF, on filtre par code INSEE
                        if transport_mode in ['METRO', 'TRAMWAY', 'RER', 'VAL']:
                            if is_in_idf(insee_code):
                                return admin_region['name'], insee_code
                        # Si c'est un train GL ou un autre type, on ne filtre pas
                        elif transport_mode == 'TRAIN':
                            return admin_region['name'], insee_code
        return None, None
    else:
        print(f"Erreur {response.status_code} lors de l'appel à l'API pour {station_name}")
        return None, None
```

```python
# Fonction pour nettoyer les noms de station
def clean_station_name(station_name):
    return station_name.strip().replace('\u200b', '').replace('\u00a0', '')

# Appliquer le nettoyage à toutes les clés du dictionnaire manual_corrections
manual_corrections_cleaned = {clean_station_name(k): v for k, v in manual_corrections.items()}

for index, row in df.iterrows():
    station_name = clean_station_name(row['nom_iv'])
    transport_mode = row['mode']  # Ex: 'METRO', 'TRAMWAY', 'GL'
    
    # Essayer de récupérer les données via l'API avec le mode de transport
    commune, insee = get_commune_from_station_sncg(station_name, api_key, transport_mode)
    
    # Si l'API ne retourne pas de résultat, utiliser les corrections manuelles si disponibles
    if not commune or not insee:
        if station_name in manual_corrections:
            commune, insee = manual_corrections[station_name]
        else:
            print(f"Aucune correspondance trouvée pour {station_name}")
    
    # Mettre à jour les colonnes du DataFrame
    df.at[index, 'Commune'] = commune
    df.at[index, 'INSEE'] = insee
```
